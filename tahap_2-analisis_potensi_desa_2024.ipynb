# Analisis Eksploratif Potensi Desa di Indonesia

## Tujuan

Tujuan dari analisis ini adalah mendapatkan insight terkait potensi desa di Indonesia dalam berbagai bidang, seperti kependudukan, fasilitas, ekonomi, dan infrastruktur.

Sumber yang digunakan yaitu data Badan Pusat Statistik (BPS) dan Portal Data Kementerian Desa (Kemendesa)

```python
# Import Libraries
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Analisis Data (EDA)

Analisis data dilakukan pada file yang telah dilakukan cleaning dalam folder data/processed

```python
# Membuat daftar file csv yang telah diunduh
script_dir = os.getcwd()
csv_files = glob.glob(os.path.join(script_dir, 'data/processed', '*.csv'))
print(len(csv_files))
```

```python
# Membuka file CSV dan disimpan dalam dataframe
dataframes = {}

for file_path in csv_files:
    file_name = os.path.splitext(os.path.basename(file_path))[0]
    df = pd.read_csv(file_path, encoding='latin1')
    dataframes[file_name] = df
```

```python
# Melihat data yang telah dibuka
for name, df in dataframes.items():
    print(f"DataFrame: {name}")
    display(df.tail())
    print("\n")
```

#### Data Karakteristik Jumlah Desa Tahun 2024

Data Karakteristik Jumlah Desa di Indonesia tahun 2024 diperoleh dengan menggabungkan file 12, 03, dan 04

```python
# Menyamakan nama provinsi antar data
peta_provinsi = {
    'Kep. Bangka Belitung': 'Kepulauan Bangka Belitung',
    'Kep. Riau': 'Kepulauan Riau',
    'Dki Jakarta': 'DKI Jakarta',
    'Di Yogyakarta': 'DI Yogyakarta'
}
```

```python
# Merge df 12 dan df 03-04 untuk melihat letak geografi
df_03 = dataframes['03_Jumlah Desa Menurut Provinsi dan Letak Geografi, 2024-processed'][:-1].copy()
df_04 = dataframes['04_Jumlah Desa yang Memiliki Fasilitas Sekolah Menurut Provinsi dan Tingkat Pendidikan, 2024-processed'][:-1].copy()
# Menyamakan nama provinsi sebagai key
df_03['Provinsi'] = df_03['Provinsi'].str.title().replace(peta_provinsi)
df_04['Provinsi'] = df_04['Provinsi'].str.title().replace(peta_provinsi)
df_12_03 = pd.merge(df_12, df_03, on='Provinsi')
# File 12, 03, dan 04 digabungkan menjadi Data Karakteristik Jumlah Desa di Indonesia tahun 2024
df_2024 = pd.merge(df_12_03, df_04, on='Provinsi')
```

```python
# Dibuat persentase untuk melihat karakteristik tiap provinsi
df_2024['Tepi Laut (%)'] = (df_2024['Tepi Laut_2024'] / df_2024['Jumlah Desa_2024']) * 100
df_2024['Bukan Tepi Laut (%)'] = (df_2024['Bukan Tepi Laut_2024'] / df_2024['Jumlah Desa_2024']) * 100
df_2024['SD (%)'] = (df_2024['SD_2024'] / df_2024['Jumlah Desa_2024']) * 100
df_2024['SMP (%)'] = (df_2024['SMP_2024'] / df_2024['Jumlah Desa_2024']) * 100
df_2024['SMU (%)'] = (df_2024['SMU_2024'] / df_2024['Jumlah Desa_2024']) * 100
df_2024['SMK (%)'] = (df_2024['SMK_2024'] / df_2024['Jumlah Desa_2024']) * 100
df_2024['Perguruan Tinggi (%)'] = (df_2024['Perguruan Tinggi_2024'] / df_2024['Jumlah Desa_2024']) * 100
```

```python
print("--Karakteristik Jumlah Desa Tahun 2024--")
df_2024.round(2)
```

```python
# Visualisasi Pie Chart untuk Persebaran Letak Geografis Desa
total_tepi_laut = df_2024['Tepi Laut_2024'].sum()
total_bukan_tepi_laut = df_2024['Bukan Tepi Laut_2024'].sum()
data_pie = pd.Series({'Tepi Laut': total_tepi_laut, 'Bukan Tepi Laut': total_bukan_tepi_laut})

sns.set_theme(style="whitegrid")
colors = sns.color_palette('pastel')[0:2]
plt.figure(figsize=(8, 8))
plt.pie(data_pie,                  
        labels=data_pie.index,     
        colors=colors,
        autopct='%1.2f%%',         
        startangle=90,
        explode=[0.02, 0.02]       # Sedikit jarak antar potongan
       )
plt.title('Proporsi Jumlah Desa Tepi Laut dan Bukan Tepi Laut di Indonesia Tahun 2024', fontsize=16)
plt.axis('equal')

plt.show()
```

```python
# Visualisasi Bar Chart untuk Perbandingan Fasilitas Sekolah dibandingkan Jumlah Desa
persentase_sd = df_2024['SD (%)'].mean()
persentase_smp = df_2024['SMP (%)'].mean()
persentase_smu = df_2024['SMU (%)'].mean()
persentase_smk = df_2024['SMK (%)'].mean()
persentase_pt = df_2024['Perguruan Tinggi (%)'].mean()

data_bar = pd.Series({'SD': persentase_sd, 'SMP': persentase_smp, 'SMU': persentase_smu, 'SMK': persentase_smk, 'PT': persentase_pt})

plt.figure(figsize=(8,8))
ax = sns.barplot(x=data_bar.index, y=data_bar.values)
plt.title("Fasilitas Sekolah yang Dimiliki Desa di Indonesia Tahun 2024 (dalam persen)")
plt.xlabel("Fasilitas")
plt.ylabel("Persentase")
plt.ylim(0,100)

for p in ax.patches:
    height = p.get_height()
    # Tempatkan teks di atas bar
    ax.text(
        x=p.get_x() + p.get_width() / 2, 
        y=height + 1,                    
        s=f'{height:.1f}',               
        ha='center'                      
    )

plt.show()
```

#### Data Tren Jumlah Desa Tahun 2019-2021

Data Tren Jumlah Desa Mandiri dan Tertinggal Tahun 2019-2021 dilakukan dengan menggabungkan file 01 dan 02

```python
# Merge file 01 untuk Jumlah Desa Mandiri
df_01a = dataframes['01_Jumlah Desa Mandiri Menurut Provinsi, 2019-processed'][:-1].copy()
df_01a['Provinsi'] = df_01a['Provinsi'].str.title().replace(peta_provinsi)
df_01b = dataframes['01_Jumlah Desa Mandiri Menurut Provinsi, 2020-processed'][:-1].copy()
df_01b['Provinsi'] = df_01b['Provinsi'].str.title().replace(peta_provinsi)
df_01c = dataframes['01_Jumlah Desa Mandiri Menurut Provinsi, 2021-processed'][:-1].copy()
df_01c['Provinsi'] = df_01c['Provinsi'].str.title().replace(peta_provinsi)
df_01ab = pd.merge(df_01a, df_01b, on='Provinsi')
df_01 = pd.merge(df_01ab, df_01c, on='Provinsi')

# Merge file 02 untuk Jumlah Desa Tertinggal
df_02a = dataframes['02_Jumlah Desa Tertinggal menurut Provinsi, 2019-processed'][:-1].copy()
df_02a['Provinsi'] = df_02a['Provinsi'].str.title().replace(peta_provinsi)
df_02b = dataframes['02_Jumlah Desa Tertinggal menurut Provinsi, 2020-processed'][:-1].copy()
df_02b['Provinsi'] = df_02b['Provinsi'].str.title().replace(peta_provinsi)
df_02c = dataframes['02_Jumlah Desa Tertinggal menurut Provinsi, 2021-processed'][:-1].copy()
df_02c['Provinsi'] = df_02c['Provinsi'].str.title().replace(peta_provinsi)
df_02ab = pd.merge(df_02a, df_02b, on='Provinsi')
df_02 = pd.merge(df_02ab, df_02c, on='Provinsi')

# Merge file 01 dan 02 untuk Data Tren Jumlah Desa
df_tren = pd.merge(df_01, df_02, on='Provinsi')
```

```python
print("-- Data Jumlah Desa Mandiri dan Tertinggal Tahun 2019-2021 --")
df_tren.round(2)
```

```python
# Visualisasi Multiple Line Chart untuk Tren Kategori Desa di Indonesia
kolom_total = ['Jumlah Desa Mandiri Menurut Provinsi (Desa)_2019', 'Jumlah Desa Mandiri Menurut Provinsi (Desa)_2020', 'Jumlah Desa Mandiri Menurut Provinsi (Desa)_2021',
               'Desa Sangat Tertinggal_2019', 'Desa Sangat Tertinggal_2020', 'Desa Sangat Tertinggal_2021',
               'Desa Tertinggal_2019', 'Desa Tertinggal_2020', 'Desa Tertinggal_2021']
total_desa_kategori = df_tren[kolom_total].sum()
df_line = total_desa_kategori.reset_index()
df_line.rename(columns={'index': 'Label', 0: 'Jumlah Desa'}, inplace=True)
df_line[['Kategori', 'Tahun']] = df_line['Label'].str.split('_', expand=True, n=1)
df_line['Kategori'] = df_line['Kategori'].replace('Jumlah Desa Mandiri Menurut Provinsi (Desa)','Desa Mandiri')
```

```python
# Buat multiple line chart
plt.figure(figsize=(12, 7))
ax = sns.lineplot(
    data=df_line,
    x='Tahun',
    y='Jumlah Desa',
    hue='Kategori', 
    marker='o',      
    linewidth=3
)
plt.title('Perbandingan Tren Jumlah Desa per Kategori', fontsize=16)
plt.xlabel('Tahun', fontsize=14)
plt.ylabel('Jumlah Desa', fontsize=14)
plt.show()
```
